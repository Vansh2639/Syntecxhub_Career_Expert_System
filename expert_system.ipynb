```python
# Career Recommendation Expert System

rules = [
    {
        "if": ["programming", "problem_solving"],
        "then": "software_engineer"
    },

    {
        "if": ["software_engineer", "mathematics"],
        "then": "ai_engineer"
    },

    {
        "if": ["creativity", "design"],
        "then": "ui_ux_designer"
    },

    {
        "if": ["communication", "marketing"],
        "then": "digital_marketer"
    },

    {
        "if": ["ai_engineer", "machine_learning"],
        "then": "ml_specialist"
    },

    {
        "if": ["software_engineer", "database"],
        "then": "backend_developer"
    }
]


def forward_chaining(facts, rules):

    inferred = set(facts)

    log = []

    changed = True

    while changed:

        changed = False

        for rule in rules:

            conditions = rule["if"]
            conclusion = rule["then"]

            # Check if all conditions are present
            if all(cond in inferred for cond in conditions):

                # Avoid duplicates
                if conclusion not in inferred:

                    inferred.add(conclusion)

                    log.append(
                        f"{' + '.join(conditions)} -> {conclusion}"
                    )

                    changed = True

    return inferred, log


print("=== Career Recommendation Expert System ===")

print("\nAvailable Skills:")
print("programming")
print("problem_solving")
print("mathematics")
print("machine_learning")
print("database")
print("creativity")
print("design")
print("communication")
print("marketing")

user_input = input(
    "\nEnter your skills separated by commas: "
)

facts = set()

for skill in user_input.split(","):
    facts.add(skill.strip().lower())

results, steps = forward_chaining(facts, rules)

print("\n===== Recommendations =====")

for item in results:

    if item not in facts:
        print("•", item)

print("\n===== Reasoning Path =====")

for step in steps:
    print(step)
```
